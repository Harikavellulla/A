```python
import pyttsx3
from IPython.display import Audio
text = "welcome to naresh tecnology data science under prekash senapthi guidence"

audio = pyttsx3.init()
audio.setProperty('rate',150)
audio.setProperty('volume',0.8)

voice = audio.getProperty('voices')

audio.say(text)

audio.save_to_file(text, 'test_male_vioce.map3')
audio.runAndWait()
```
