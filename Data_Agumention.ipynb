```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img,img_to_array

# create an image data generator with argumentation
datagen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.4,
    height_shift_range = 0.4,
    shear_range = 0.4,
    zoom_range = 0.4,
    horizontal_flip = True,
    rescale = None,
    fill_mode = 'constant'    #fill_mode = 'wrap','nearest',..
)
```

```python
img = load_img(r"c:\Users\LENOVO\OneDrive\Desktop\car.jpg")
img
```

```python
# covert to numpy array
x = img_to_array(img)
x = x.reshape((1,) + x.shape)   # reshape to (1,height,width,channels)

# Generate aumentted images and saves
i = 0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir = r"c:\Users\LENOVO\OneDrive\Desktop\Data",
                          save_prefix = 'car',
                          save_format = 'jpeg'):
    i += 1
    if i > 18:
        break   # prevent infinite loop
```
